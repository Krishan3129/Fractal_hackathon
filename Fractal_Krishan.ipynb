```python

# Create DataFrame
df = pd.DataFrame(data)

# Function to extract numeric value and classification for pH
def extract_ph_info(text):
    numeric_match = re.search(r"\b\d+\.\d+\b", text)
    numeric_value = float(numeric_match.group()) if numeric_match else None
    if "acidic" in text.lower():
        description = "0"
    elif "neutral" in text.lower():
        description = "1"
    elif "alkaline" in text.lower():
        description = "2"
    else:
        description = "4"
    return pd.Series([numeric_value, description])

# Function to extract numeric value and classification for Hardness
def extract_hardness_info(text):
    numeric_match = re.search(r"\b\d+\.\d+\b", text)
    numeric_value = float(numeric_match.group()) if numeric_match else None
    if "soft" in text.lower():
        description = "0"
    elif "moderate" in text.lower():
        description = "1"
    elif "hard" in text.lower():
        description = "2"
    elif "moderately hard" in text.lower():
        description = "3"
    else:
        description = "4"
    return pd.Series([numeric_value, description])

# Function to extract numeric value and classification for Solids
def extract_solids_info(text):
    numeric_match = re.search(r"\b\d+\.\d+\b", text)
    numeric_value = float(numeric_match.group()) if numeric_match else None
    if "low" in text.lower():
        description = "0"
    elif "moderate" in text.lower():
        description = "1"
    elif "significant" in text.lower():
        description = "2"
    else:
        description = "3"
    return pd.Series([numeric_value, description])

# Function to extract numeric value and classification for Turbidity
def extract_turbidity_info(text):
    numeric_match = re.search(r"\b\d+\.\d+\b", text)
    numeric_value = float(numeric_match.group()) if numeric_match else None
    if "cloudy" in text.lower():
        description = "0"
    elif "clear" in text.lower():
        description = "1"
    elif "suspended particles" in text.lower():
        description = "2"
    else:
        description = "3"
    return pd.Series([numeric_value, description])

# Apply the extraction functions to each column and create new columns
df[['pH_value', 'pH_description']] = df['ph'].apply(extract_ph_info)
df[['Hardness_value', 'Hardness_description']] = df['Hardness'].apply(extract_hardness_info)
df[['Solids_value', 'Solids_description']] = df['Solids'].apply(extract_solids_info)
df[['Turbidity_value', 'Turbidity_description']] = df['Turbidity'].apply(extract_turbidity_info)

# Display the updated DataFrame
print(df)
```
